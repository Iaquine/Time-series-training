```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from statsmodels.datasets import co2, get_rdataset
```

```python
file = Path('/Users/iaquine/Documents/GitHub/Time-series-training/datasets/closing_price.csv')

#dataset 1: Contém os preços da açoes da IBM, Apple e Microsoft de Novembro de 2019 a Novembro de 2021 
closing_price = pd.read_csv(file, 
                            index_col='Date',
                            parse_dates=True)


#dataset 2: Contém niveis de dioxido de carbono medidos semanalmente em Mauna Loa
co2_df = co2.load_pandas().data
#co2_df = co2_df.ffill() #Essa linha preenche os valores vazios com os valores da coluna anterior
```

```python
#dataset 3: Contém o número de passageiros mensais de 1949 a 1960
air_passengers = get_rdataset("AirPassengers")
airp_df = air_passengers.data
airp_df.index = pd.date_range('1949', '1961', freq='M')
airp_df.drop(columns=['time'], inplace=True)

# store for future use
airp_df.reset_index().rename(columns={'value':'passengers', 'index':'date'}).to_csv('/Users/iaquine/Documents/GitHub/Time-series-training/datasets/air_passenger.csv', index=False)
```
